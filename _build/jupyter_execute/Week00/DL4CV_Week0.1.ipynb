# Activation Functions for Deep Learning
## Machine Learning & Computer Vision Prerequisites

Activation functions introduce **non-linearity** into neural networks. Without them, a deep neural network behaves like a single linear model regardless of the number of layers.

---

# Why Do We Need Activation Functions?

Suppose a neuron computes:

$$
z = wx + b
$$

Without an activation function:

$$
y = z
$$

This is simply a linear equation.

By applying an activation function:

$$
y = f(z)
$$

the network can learn complex nonlinear relationships.

---

# 1. Binary Step Function

## Formula

$$
f(x)=
\begin{cases}
1, & x \ge 0\\
0, & x < 0
\end{cases}
$$

## Example

Find output for:

$$
x = 2
$$

Since

$$
2 \ge 0
$$

Output:

$$
f(2)=1
$$

---

Find output for:

$$
x=-3
$$

Since

$$
-3 < 0
$$

Output:

$$
f(-3)=0
$$

## Advantages

- Very simple
- Useful in perceptrons

## Disadvantages

- Not differentiable
- Cannot be used in backpropagation

---

# 2. Linear Activation

## Formula

$$
f(x)=x
$$

## Example

Input:

$$
x=5
$$

Output:

$$
f(5)=5
$$

## Derivative

$$
f'(x)=1
$$

## Problem

No nonlinearity.

Deep network becomes equivalent to a single layer.

---

# 3. Sigmoid Function

One of the most important activation functions.

## Formula

$$
\sigma(x)=
\frac{1}{1+e^{-x}}
$$

## Output Range

$$
0 < \sigma(x) < 1
$$

Useful for probability prediction.

---

## Example 1

Input:

$$
x=0
$$

Substitute:

$$
\sigma(0)
=
\frac{1}{1+e^{0}}
$$

Since

$$
e^0=1
$$

$$
\sigma(0)
=
\frac{1}{2}
=
0.5
$$

### Answer

$$
\boxed{0.5}
$$

---

## Example 2

Input:

$$
x=2
$$

$$
\sigma(2)
=
\frac{1}{1+e^{-2}}
$$

$$
=
\frac{1}{1+0.1353}
$$

$$
=
0.8808
$$

### Answer

$$
\boxed{0.8808}
$$

---

## Derivative

A famous result:

$$
\sigma'(x)
=
\sigma(x)(1-\sigma(x))
$$

---

## Example

If

$$
\sigma(x)=0.8
$$

Then

$$
\sigma'(x)
=
0.8(1-0.8)
$$

$$
=
0.16
$$

### Answer

$$
\boxed{0.16}
$$

---

## Advantages

- Output interpretable as probability
- Smooth curve

## Disadvantages

- Vanishing gradient problem
- Slow training

---

# 4. Tanh Function

## Formula

$$
\tanh(x)
=
\frac{e^x-e^{-x}}
{e^x+e^{-x}}
$$

## Output Range

$$
-1 < \tanh(x) < 1
$$

---

## Example

Input:

$$
x=0
$$

$$
\tanh(0)=0
$$

### Answer

$$
\boxed{0}
$$

---

## Example

Input:

$$
x=1
$$

$$
\tanh(1)
=
0.7616
$$

### Answer

$$
\boxed{0.7616}
$$

---

## Derivative

$$
\tanh'(x)
=
1-\tanh^2(x)
$$

---

## Example

If

$$
\tanh(x)=0.7616
$$

Then

$$
\tanh'(x)
=
1-(0.7616)^2
$$

$$
=
0.42
$$

### Answer

$$
\boxed{0.42}
$$

---

## Advantages

- Zero-centered output
- Better than Sigmoid

## Disadvantages

- Still suffers from vanishing gradients

---

# 5. ReLU (Rectified Linear Unit)

Most widely used activation function.

## Formula

$$
f(x)=\max(0,x)
$$

or

$$
f(x)=
\begin{cases}
0, & x<0\\
x, & x\ge0
\end{cases}
$$

---

## Example 1

Input:

$$
x=5
$$

Output:

$$
f(5)=5
$$

---

## Example 2

Input:

$$
x=-4
$$

Output:

$$
f(-4)=0
$$

---

## Derivative

$$
f'(x)=
\begin{cases}
0, & x<0\\
1, & x>0
\end{cases}
$$

---

## Example

Input:

$$
x=3
$$

Derivative:

$$
f'(3)=1
$$

### Answer

$$
\boxed{1}
$$

---

## Advantages

- Fast computation
- Reduces vanishing gradients
- Most popular hidden layer activation

## Disadvantages

- Dead ReLU problem

---

# 6. Leaky ReLU

Improves ReLU.

## Formula

$$
f(x)=
\begin{cases}
x,& x>0\\
\alpha x,& x<0
\end{cases}
$$

Typically

$$
\alpha=0.01
$$

---

## Example

Input:

$$
x=-5
$$

$$
f(-5)
=
0.01(-5)
$$

$$
=
-0.05
$$

### Answer

$$
\boxed{-0.05}
$$

---

## Derivative

$$
f'(x)=
\begin{cases}
1,& x>0\\
0.01,& x<0
\end{cases}
$$

---

# 7. Softmax Function

Used in multi-class classification.

## Formula

For class i:

$$
P_i
=
\frac{e^{z_i}}
{\sum_{j=1}^{n}e^{z_j}}
$$

---

## Example

Suppose logits are:

$$
[2,1,0]
$$

### Step 1

Compute exponentials:

$$
e^2=7.389
$$

$$
e^1=2.718
$$

$$
e^0=1
$$

### Step 2

Sum:

$$
7.389+2.718+1
=
11.107
$$

### Step 3

Probabilities

Class 1:

$$
\frac{7.389}{11.107}
=
0.665
$$

Class 2:

$$
\frac{2.718}{11.107}
=
0.245
$$

Class 3:

$$
\frac{1}{11.107}
=
0.090
$$

### Final Output

$$
[0.665,\;0.245,\;0.090]
$$

Check:

$$
0.665+0.245+0.090
=
1
$$

---

# Activation Function Comparison

| Function | Range | Differentiable | Typical Use |
|-----------|---------|---------|---------|
| Binary Step | 0,1 | No | Perceptron |
| Linear | (-∞,∞) | Yes | Regression Output |
| Sigmoid | (0,1) | Yes | Binary Classification |
| Tanh | (-1,1) | Yes | Hidden Layers (Older Networks) |
| ReLU | [0,∞) | Yes | Hidden Layers |
| Leaky ReLU | (-∞,∞) | Yes | Hidden Layers |
| Softmax | (0,1) | Yes | Multi-class Output |

---

# Which Activation Function Should I Use?

## Hidden Layers

Use:

$$
\boxed{\text{ReLU}}
$$

or

$$
\boxed{\text{Leaky ReLU}}
$$

---

## Binary Classification Output

Use:

$$
\boxed{\text{Sigmoid}}
$$

Example:

- Cat vs Dog
- Spam vs Not Spam

---

## Multi-Class Classification Output

Use:

$$
\boxed{\text{Softmax}}
$$

Example:

- Digits 0-9
- ImageNet Classes
- Object Recognition

---

# Interview Questions

### Why is ReLU popular?

- Computationally efficient
- Less vanishing gradient
- Faster convergence

### Why not use Sigmoid in hidden layers?

Because gradients become very small for large positive or negative values.

### Why use Softmax?

Converts raw scores into probabilities that sum to 1.

### Which activation is most common in modern CNNs?

$$
\boxed{\text{ReLU}}
$$

Used in:

- AlexNet
- VGG
- ResNet
- YOLO
- EfficientNet
