#### Derivative Rules

##### Constant Multiple Rule

$$\frac{d}{dx}[c f(x)] = c f'(x)$$

##### Sum Rule

$$\frac{d}{dx}[f(x)+g(x)] = f'(x)+g'(x)$$

##### Difference Rule

$$\frac{d}{dx}[f(x)-g(x)] = f'(x)-g'(x)$$

##### Product Rule

$$\frac{d}{dx}[f(x)g(x)] = f'(x)g(x)+f(x)g'(x)$$

##### Quotient Rule


$$\frac{d}{dx}\left[\frac{f(x)}{g(x)}\right] = \frac{g(x)f'(x)-f(x)g'(x)} {(g(x))^2}$$

##### Chain Rule

$$\frac{d}{dx}[f(g(x))] = f'(g(x))\cdot g'(x)$$

### Example 1: Polynomial Function

**Problem:** Find the derivative of $f(x)=3x^4+2x^2-5x+7$

**Solution**

$\begin{aligned}
f'(x)
&= \frac{d}{dx}(3x^4+2x^2-5x+7) \\
&= \frac{d}{dx}(3x^4)+\frac{d}{dx}(2x^2)+\frac{d}{dx}(-5x)+\frac{d}{dx}(7) \\
&= 12x^3+4x-5+0 \\
&= 12x^3+4x-5
\end{aligned}$

**Answer:** $\boxed{f'(x)=12x^3+4x-5}$

### Example 2: Chain Rule

#### Problem

Find

$$
\frac{d}{dx}(x^2+1)^5
$$

#### Solution

Let

$$
u=x^2+1
$$

Then

$$
f(x)=u^5
$$

Differentiate outer function first.

$$
\frac{d}{du}(u^5)=5u^4
$$

Differentiate inner function.

$$
\frac{du}{dx}=2x
$$

Apply Chain Rule.

$$
\frac{df}{dx}
=
5u^4 \cdot 2x
$$

Substitute back.

$$
\boxed{
\frac{d}{dx}(x^2+1)^5
=
10x(x^2+1)^4
}
$$

### Example 3: Exponential Function

#### Problem

Find

$$
\frac{d}{dx}(e^{3x})
$$

#### Solution

Use

$$
\frac{d}{dx}(e^u)=e^u\frac{du}{dx}
$$

where

$$
u=3x
$$

Then

$$
\frac{du}{dx}=3
$$

Therefore

$$
\boxed{
\frac{d}{dx}(e^{3x})
=
3e^{3x}
}
$$

### Example 4: Logarithmic Function

#### Problem

Find

$$
\frac{d}{dx}\ln(x^2+1)
$$

#### Solution

Use

$$
\frac{d}{dx}\ln(u)
=
\frac{u'}{u}
$$

Let

$$
u=x^2+1
$$

Then

$$
u'=2x
$$

Substitute.

$$
\boxed{
\frac{d}{dx}\ln(x^2+1)
=
\frac{2x}{x^2+1}
}
$$

### Example 5: Sigmoid Function

#### Problem

Find derivative of

$$
\sigma(x)
=
\frac{1}{1+e^{-x}}
$$

#### Solution

Rewrite as

$$
\sigma(x)
=
(1+e^{-x})^{-1}
$$

Differentiate.

$$
\sigma'(x)
=
-(1+e^{-x})^{-2}
(-e^{-x})
$$

Simplify.

$$
\sigma'(x)
=
\frac{e^{-x}}
{(1+e^{-x})^2}
$$

Using

$$
\sigma(x)=\frac{1}{1+e^{-x}}
$$

we obtain

$$
\boxed{
\sigma'(x)
=
\sigma(x)(1-\sigma(x))
}
$$

This is one of the most important derivatives in Deep Learning.

### Example 6: Mean Squared Error (MSE)

#### Problem

Given

$$
L=(y-\hat y)^2
$$

find

$$
\frac{dL}{d\hat y}
$$

#### Solution

Let

$$
u=y-\hat y
$$

Then

$$
L=u^2
$$

Differentiate outer function.

$$
\frac{dL}{du}=2u
$$

Differentiate inner function.

$$
\frac{du}{d\hat y}=-1
$$

Apply Chain Rule.

$$
\frac{dL}{d\hat y}
=
2u(-1)
$$

Substitute back.

$$
\boxed{
\frac{dL}{d\hat y}
=
-2(y-\hat y)
}
$$

Equivalent form:

$$
\boxed{
\frac{dL}{d\hat y}
=
2(\hat y-y)
}
$$

Used in Linear Regression and Neural Networks.

### Example 7: Linear Regression Gradient

#### Problem

Given

$$
\hat y=wx+b
$$

and

$$
L=(y-\hat y)^2
$$

Find

$$
\frac{dL}{dw}
$$

#### Solution

Substitute prediction.

$$
L=(y-(wx+b))^2
$$

Let

$$
u=y-(wx+b)
$$

Then

$$
L=u^2
$$

Differentiate.

$$
\frac{dL}{du}=2u
$$

and

$$
\frac{du}{dw}=-x
$$

Apply Chain Rule.

$$
\frac{dL}{dw}
=
2u(-x)
$$

Substitute back.

$$
\boxed{
\frac{dL}{dw}
=
-2x(y-\hat y)
}
$$

This gradient is used in Gradient Descent.

### Example 8: Softmax Cross Entropy (Simplified)

#### Problem

Given

$$
L=-\log(p)
$$

Find

$$
\frac{dL}{dp}
$$

#### Solution

Use logarithmic derivative rule.

$$
\frac{d}{dp}\ln(p)
=
\frac{1}{p}
$$

Therefore

$$
\boxed{
\frac{dL}{dp}
=
-\frac{1}{p}
}
$$

Used in Classification and Computer Vision tasks.

### Key ML Derivatives Summary

$$
\frac{d}{dx}(x^n)=nx^{n-1}
$$

$$
\frac{d}{dx}(e^x)=e^x
$$

$$
\frac{d}{dx}(\ln x)=\frac1x
$$

$$
\frac{d}{dx}\sin x=\cos x
$$

$$
\frac{d}{dx}\cos x=-\sin x
$$

$$
\sigma'(x)=\sigma(x)(1-\sigma(x))
$$

$$
\frac{d}{d\hat y}(y-\hat y)^2
=
2(\hat y-y)
$$

$$
\frac{dL}{dw}
=
-2x(y-\hat y)
$$

These derivatives alone cover roughly 80–90% of the mathematics encountered in Linear Regression, Logistic Regression, Neural Networks, Backpropagation, Deep Learning, and Computer Vision.
