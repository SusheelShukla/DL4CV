### Derivative Rules

- **Constant Multiple Rule:** $\frac{d}{dx}[c f(x)] = c f'(x)$

- **Sum Rule:** $\frac{d}{dx}[f(x)+g(x)] = f'(x)+g'(x)$

- **Difference Rule:** $\frac{d}{dx}[f(x)-g(x)] = f'(x)-g'(x)$

- **Product Rule:** $\frac{d}{dx}[f(x)g(x)] = f'(x)g(x)+f(x)g'(x)$

- **Quotient Rule:** $\frac{d}{dx}\left[\frac{f(x)}{g(x)}\right] = \frac{g(x)f'(x)-f(x)g'(x)} {(g(x))^2}$

- **Chain Rule:** $\frac{d}{dx}[f(g(x))] = f'(g(x))\cdot g'(x)$

## Derivatives for Machine Learning and Computer Vision

### Example 1: Polynomial Function

**Problem:** Find the derivative of $f(x)=3x^4+2x^2-5x+7$

**Solution**

$\begin{aligned}
f'(x)
&= \frac{d}{dx}(3x^4+2x^2-5x+7) \\
&= \frac{d}{dx}(3x^4)+\frac{d}{dx}(2x^2)+\frac{d}{dx}(-5x)+\frac{d}{dx}(7) \\
&= 12x^3+4x-5+0 \\
&= 12x^3+4x-5
\end{aligned}$

**Answer:** $\boxed{f'(x)=12x^3+4x-5}$

---


### Example 2: Chain Rule

Find

$
\frac{d}{dx}(x^2+1)^5
$

**Solution:**

$\begin{aligned}
&\text{Let u} = x^2+1, \\
& Then f(x) = u^5 \\
& \text{Differentiate outer function first.} \\
& \frac{d}{du}(u^5) = 5u^4 \\
& \text{Differentiate inner function.} \\
& \frac{du}{dx} = 2x \\
& \text{Apply Chain Rule.} \\
& \frac{df}{dx} = 5u^4 \cdot 2x \\
& \text{Substitute back.} \\
& \boxed{\frac{d}{dx}(x^2+1)^5 = 10x(x^2+1)^4} \\
\end{aligned}$

### Example 3: Exponential Function

Find

$
\frac{d}{dx}(e^{3x})
$

### Solution

$\begin{aligned}
& \text{Use } \frac{d}{dx}(e^u)=e^u\frac{du}{dx} \\
& \text{Let } u = 3x \\
& \frac{du}{dx} = 3 \\
& \text{Apply the Chain Rule.} \\
& \frac{d}{dx}(e^{3x}) = e^u \cdot 3 \\
& \text{Substitute back.} \\
& \boxed{\frac{d}{dx}(e^{3x}) = 3e^{3x}} \\
\end{aligned}$

---



### Example 4: Logarithmic Function

Find

$
\frac{d}{dx}\ln(x^2+1)
$

### Solution

$\begin{aligned}
& \text{Use } \frac{d}{dx}\ln(u)=\frac{u'}{u} \\
& \text{Let } u = x^2+1 \\
& \frac{du}{dx} = 2x \\
& \text{Substitute into the formula.} \\
& \boxed{\frac{d}{dx}\ln(x^2+1)=\frac{2x}{x^2+1}} \\
\end{aligned}$

---



### Example 5: Sigmoid Function

Find derivative of

$
\sigma(x)=\frac{1}{1+e^{-x}}
$

### Solution

$\begin{aligned}
& \text{Rewrite as } \sigma(x)=(1+e^{-x})^{-1} \\
& \text{Differentiate using the Chain Rule.} \\
& \sigma'(x)=-(1+e^{-x})^{-2}(-e^{-x}) \\
& \text{Simplify.} \\
& \sigma'(x)=\frac{e^{-x}}{(1+e^{-x})^2} \\
& \text{Express in terms of } \sigma(x). \\
& \boxed{\sigma'(x)=\sigma(x)(1-\sigma(x))} \\
\end{aligned}$

### Example 6: Mean Squared Error (MSE)

Find

$
\frac{dL}{d\hat y}
$

given

$
L=(y-\hat y)^2
$

### Solution

$\begin{aligned}
& \text{Let } u = y-\hat y \\
& \text{Then } L=u^2 \\
& \text{Differentiate outer function first.} \\
& \frac{dL}{du}=2u \\
& \text{Differentiate inner function.} \\
& \frac{du}{d\hat y}=-1 \\
& \text{Apply the Chain Rule.} \\
& \frac{dL}{d\hat y}=2u(-1) \\
& \text{Substitute back.} \\
& \boxed{\frac{dL}{d\hat y}=-2(y-\hat y)=2(\hat y-y)} \\
\end{aligned}$

### Example 7: Linear Regression Gradient

Find

$
\frac{dL}{dw}
$

given

$
\hat y=wx+b
$

and

$
L=(y-\hat y)^2
$

### Solution

$\begin{aligned}
& \text{Substitute } \hat y=wx+b \\
& L=(y-(wx+b))^2 \\
& \text{Let } u=y-(wx+b) \\
& \text{Then } L=u^2 \\
& \text{Differentiate outer function first.} \\
& \frac{dL}{du}=2u \\
& \text{Differentiate inner function.} \\
& \frac{du}{dw}=-x \\
& \text{Apply the Chain Rule.} \\
& \frac{dL}{dw}=2u(-x) \\
& \text{Substitute back.} \\
& \boxed{\frac{dL}{dw}=-2x(y-\hat y)} \\
\end{aligned}$

### Example 8: Softmax Cross Entropy (Simplified)

Find

$
\frac{dL}{dp}
$

given

$
L=-\log(p)
$

### Solution

$\begin{aligned}
& \text{Use } \frac{d}{dp}\ln(p)=\frac{1}{p} \\
& \text{Differentiate.} \\
& \frac{dL}{dp}=-\frac{1}{p} \\
& \boxed{\frac{dL}{dp}=-\frac{1}{p}} \\
\end{aligned}$
